## 適応的棄却サンプリング

凹な確率密度関数の対数 $\log p(z)$ を持つ分布が多いことの確認

```python
xx = np.linspace(-5, 5)
xxp = np.linspace(0, 5)

fig, axes = plt.subplots(2,2, figsize=(6,5))

axes[0,0].plot(xx, stats.norm(0,1).logpdf(xx), label='$\\log\\mathcal{N}(0,1)$')
axes[0,1].plot(xxp, stats.gamma(2).logpdf(xxp), label='$\\log\\mathrm{Gamma}(2)$')
axes[1,0].plot(xxp, stats.expon(3).logpdf(xxp), label='$\\log\\mathrm{Exp}(3)$')
axes[1,1].plot(xxp, stats.beta(3,5).logpdf(xxp), label='$\\log\\mathrm{Beta}(3, 4)$')

for ax in fig.get_axes():
    ax.legend(loc='upper right')
```
